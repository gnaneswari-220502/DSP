```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Customer Purchasing Behaviors.csv")
```

```python
print("Shape of dataset (rows, columns):")
print(df.shape)
```

```python
print("\nColumn Names:")
print(df.columns)
```

```python
print("\nFirst 5 rows:")
print(df.head())
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
print("\nMissing values after filling:")
print(df.isnull().sum()
```

```python
print("\nMissing values after filling:")
print(df.isnull().sum())
```

```python
print("\nMedian values:")
print(df[num_cols].median())
```

```python
print("\nStandard deviation:")
print(df[num_cols].std())
```

```python
print("\nMinimum values:")
print(df[num_cols].min())

print("\nMaximum values:")
print(df[num_cols].max())
```
